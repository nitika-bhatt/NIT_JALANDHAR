```python
student_attendance = {}

def add_present(student):
    if student in student_attendance:
        student_attendance[student] = student_attendance[student] + 1
    else:
        student_attendance[student] = 1
    print("Attendance is marked for the student: ", student)

def show_attendance():
    print("Student Attendance Details are as follows: ")
    for student in student_attendance:
        print(student, "->", student_attendance[student])

def clear_attendance():
    student_attendance = {}
    print("All attendance records cleared")

while True:
    print("Choose an option")
    print("1. Add Attendance")
    print("2. Display Attendance")
    print("3. Reset Attendance")
    print("4. Exit")

    option = int(input("Enter option number: "))

    if option == 1:
        student_name = input("Enter student name: ")
        add_present(student_name)

    elif option == 2:
        show_attendance()

    elif option == 3:
        student_attendance.clear()
        print("Attendance reset successfully")

    elif option == 4:
        print("Program terminated")
        break

    else:
        print("Please enter a valid option")
```

```python
def sum_proper_divisors(n):
    if n == 1:
        return 0

    s = 1  
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            s += i
            if i != n // i:
                s += n // i
    return s


limit = int(input("Enter a number: "))
total = 0

for a in range(2, limit):
    b = sum_proper_divisors(a)

    
    if b != a and b < limit:
        if sum_proper_divisors(b) == a:
            total += a

print("Sum of amicable numbers under", limit, "is:", total)
```
