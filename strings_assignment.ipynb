```python
# Program to encrypt a message by adding a key value to every character

message = input("Enter the message: ")
key = int(input("Enter the key value: "))

encrypted = ""

for ch in message:
    if ch != " ":
        encrypted += chr(ord(ch) + key)
    else:
        encrypted += " "

print("Encrypted Message:", encrypted)
```

```python
# Program to check if a given character is present in a string or not
# and print its index without using find()

string = input("Enter a string: ")
ch = input("Enter the character to search: ")

found = False  # To track if character is found

for i in range(len(string)):
    if string[i] == ch:
        print(f"Character '{ch}' found at index {i}")
        found = True
        break

if not found:
    print(f"Character '{ch}' not found in the string.")


```

```python
# Program to get a string made up of 4 copies of the last 2 characters

string = input("Enter a string (at least 2 characters long): ")

if len(string) >= 2:
    last2 = string[-2:]       # Get last two characters
    result = last2 * 4        # Repeat them 4 times
    print("Resulting string:", result)
else:
    print("String length should be at least 2.")
```

```python
# Program to reverse a string if its length is a multiple of 4

string = input("Enter a string: ")

if len(string) % 4 == 0:
    result = string[::-1]   # Reverse the string
    print("Reversed string:", result)
else:
    print("Length is not a multiple of 4, so no change:", string)
```

```python
# wap to print onm ly first 2 chars of every word, do it using resgular expression and traditional looking
text = input("Enter a sentence: ")

words = text.split()
result = []

for word in words:
    result.append(word[:2])

print("First 2 chars of every word (traditional):")
print(result)
```

```python
#using regular expression
import re

text = input("Enter a sentence: ")

# \b means word boundary, \w{2} means first two word characters
first_two = re.findall(r'\b\w{2}', text)

print("First 2 chars of every word (using regex):")
print(first_two)
```

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
found = False

for ch in text:
    if ch in vowels:
        found = True
        break

if found:
    print("The string has at least one vowel.")
else:
    print("The string has no vowels.")
```
