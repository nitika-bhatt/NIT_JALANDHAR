```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

# Creating objects
s1 = Student("Rahul", 85)
s2 = Student("Anita", 90)

# Using overloaded + operator
total_marks = s1 + s2

print("Total Marks:", total_marks)
```
